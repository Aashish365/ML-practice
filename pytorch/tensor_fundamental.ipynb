```python
# Creating a tensor 

# Empty tensor contains random values
empty_tensor=torch.empty(3,2)
print(empty_tensor)

# Zeros tensor
zeros_tensor=torch.zeros(3,3)
print(zeros_tensor)

# Ones Tensor
ones_tensor=torch.ones(3,2)
print(ones_tensor)

# Fill tensor
full_tensor=torch.full([3,2],4,dtype=torch.float16)
print(full_tensor)

# arange(start,end,step)
range_tensor=torch.arange(1,30,5)
print(rand_tensor)

# Linspace (start,end,step): uniform space between numbers
linspace_tensor=torch.linspace(1,20,30)
print(linspace_tensor)

# like another matrix : with same shape as another matrix
sample=torch.rand(3,5)
my_zeros=torch.zeros_like(sample # Zeros tensor with same shape of another tensor

print(sample.shape,my_zeros.shape)

```

```python
# Attributes of tensors:
data=[[1,2,5,6,2,32],[3,4,6,7,9,34]]
data_tensor=torch.tensor(data,dtype=torch.float32)

print(f"Shape: {data_tensor.shape} Size:{data_tensor.size()}")
print(f"Dimension: {data_tensor.ndim} ")
print(f"No of elements: {data_tensor.numel()}")
print(f"Flatten : {torch.flatten(data_tensor)}")

print(f"Reshape : {data_tensor.reshape(3,4)}")
print(f"Data Type: {data_tensor.dtype}")
print(f"Device: {data_tensor.device}")

```
