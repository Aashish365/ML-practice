```python
import numpy as np 
import matplotlib.pyplot as plt 
import torch
import torch.nn as nn 
from torch.utils.data import Dataset,DataLoader

from torchvision import datasets,transforms
from torchinfo import summary

torch.manual_seed(43)
np.random.seed(43)
```

```python
# device declaration
device="cuda" if torch.cuda.is_available() else "cpu"
```

```python
# Data pipeline for inbuilt data 

# Define a transform
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,)) # Mean and std for MNIST
])

# PUlling data from inbuilt datasets of torchvision module
train_datasets=datasets.MNIST(root="./data",train=True,download=True,transform=transform)
test_datasets=datasets.MNIST(root="./data",train=False,download=True,transform=transform)

# create data loaders 
train_dataloader=DataLoader(train_datasets,batch_size=64,shuffle=True)
test_dataloader=DataLoader(test_datasets,batch_size=64,shuffle=False)

classes=torch.unique(train_datasets.targets)

```

```python
# visualize some samples from the data.
sample_size=9
random_indices=np.random.randint(0,len(train_datasets),size=sample_size)


rows=3
cols=3
fig,ax=plt.subplots(3,3)
for i in range(rows):
    for j in range(cols):
        idx=random_indices[i*cols + j]
        img,label=train_datasets[idx]

        ax[i,j].imshow(img.squeeze().numpy(),cmap="gray")
        ax[i,j].set_title(label)
        ax[i,j].axis('off')
        
plt.tight_layout()
plt.show()
```

```python
# Defining a model 

class MNIST_Classifier(nn.Module):
    def __init__(self,input_units,hidden_units,num_classes):
        super().__init__()

          # Convolutional layers
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)   # 28x28 → 28x28
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)  # 28x28 → 28x28
        self.pool = nn.MaxPool2d(2, 2) 
        self.relu=nn.ReLU()
        
         # Fully connected layers
        self.fc1 = nn.Linear(64 * 14 * 14, 128)
        self.fc2 = nn.Linear(128, num_classes)
       

    def forward(self,x):
         # Conv + ReLU + Pool
        x = self.relu(self.conv1(x))
        x = self.pool(self.relu(self.conv2(x)))
        
        # Flatten
        x = x.view(x.size(0), -1)
        
        # Fully connected
        x = self.relu(self.fc1(x))
        x = self.fc2(x)   # logits, no softmax needed
        return x



# Object initialization
classifier=MNIST_Classifier(28*28,5,len(classes))
summary(classifier,input_size=(1,1,28,28))


# pulling a data from dataloader and testing it in our classifier object.
images,labels=next(iter(train_dataloader))
classifier(images[0].unsqueeze(0).to(device))
```

```python
# setting up training components
learning_rate=0.01

# loss_function cross entropy loss fn
loss_fn=nn.CrossEntropyLoss()

# choosing optimizer 
optimizer=torch.optim.Adam(classifier.parameters(),lr=learning_rate)

# accuracy function 
def accuracy_fn(logits, labels):
    predicted_classes = logits.argmax(dim=1)
    correct = (predicted_classes == labels).sum().item()
    total = labels.size(0)
    return correct / total  
   
```

```python
# Training Loop
epochs=5
losses=[]
accuracies=[]

for epoch in range(1,epochs+1):
    epoch_loss=0
    epoch_acc=0
    
    classifier.train()
    total_batches=len(train_dataloader)
    
    for images,labels in train_dataloader: 
        images,labels=images.to(device),labels.to(device)

        # Forward Pass
        logits=classifier(images) 
        loss=loss_fn(logits,labels) 
        acc=accuracy_fn(logits,labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step() 

        epoch_loss+=loss.item()
        epoch_acc+=acc
        
    avg_loss = epoch_loss / total_batches
    avg_acc = epoch_acc / total_batches
    losses.append(avg_loss)
    accuracies.append(avg_acc)
    print(f"Epoch [{epoch}/{epochs}]  "
          f"Loss: {avg_loss:.4f}  "
          f"Acc:  {avg_acc:.4f}")
   
    
        
```

```python
classifier.eval()
test_loss = 0
test_acc = 0

with torch.no_grad():
    for images, labels in test_dataloader:
        images = images.to(device)
        labels = labels.to(device)

        logits = classifier(images)

        test_loss += loss_fn(logits, labels).item()
        test_acc += accuracy_fn(logits, labels)

test_loss /= len(test_dataloader)
test_acc /= len(test_dataloader)

print(f"Test Loss: {test_loss:.4f}  Test Acc: {test_acc:.4f}")
```
